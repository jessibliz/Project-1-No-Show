## Introduction

### Dataset Description 

 This Dataset contains information from 100k medical appointments in Brazil. The Dataset contains a total of 110526 rows and 14 columns with headings as follows: 
PatientId: identifies the Identity number of each Patient.
AppointmentID: shows the appointment number of each Patient.
Gender: classifies Patients sex as either Male or a Female.
ScheduledDay: shows the Date of appointment and Time
AppointmentDay: shows only the Date of Appointment
Age: gives information about the Age of individual Patient.
Neigbourhood: indicates the location of the Facilility
Scholarship: shows Patient under any form of Scholarship programm
Hipertension: shows Patient having Hipertension.
Diabetes: identifies Patient suffering from Diabetes.
Alcoholism: shows Patient who are Alcoholic
Handcap: shows having Physical chanlleges. 
SMS_received: gives information about Patient that received SMS or not
No-show: indicates Patient that showed up for their appointment.


### Question(s) for Analysis

Could factors such as Scholarship,Age,SMS_received and Hipertension affect why Patient didn't show up for their Scheduled-Appountment?

# Comparism of multiple independent variable and dependent variable on different chat.

The codes below shows the count of scholarship and SMS.

```python
SMS_count = df_No_show['SMS_received'].value_counts()
SMS_count
```

```python
Scholarship_count = df_No_show['Scholarship'].value_counts()
Scholarship_count
```

The cell below showed the value_counts between Scholarship and SMS_received. This helps to for better Visualisation and representation of multiple variables(Scholarship and SMS_received) against an dependent variable(No_show)

```python
df_No_show['Scholarship'].value_counts().plot(kind='bar', alpha=0.5,color='blue',label ='Scholarship')
df_No_show['SMS_received'].value_counts().plot(kind='bar', alpha=0.5,color='orange',label ='SMS_received')
plt.title('Comparism of SMS_received & Scholarship value_counts')
plt.xlabel('SMS_received & Scholarship')
plt.ylabel('Number of Patients')
plt.legend();
```
